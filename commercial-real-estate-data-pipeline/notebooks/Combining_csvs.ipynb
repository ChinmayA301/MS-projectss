```python
import pandas as pd
import glob

# Getting a list of all CSV files in the current directory
all_files = glob.glob('cgiplus*.csv')

# Listing to hold dataframes
dataframes = []

for file in all_files:
    # Reading each CSV file into a dataframe (as I have multiple cgiplus files)
    df = pd.read_csv(file, index_col=None, header=0)
    dataframes.append(df)

# Combining all dataframes into a single dataframe
combined_df = pd.concat(dataframes, axis=0, ignore_index=True)

# Saving the combined dataframe to a new CSV file
combined_df.to_csv('combined_cgiplus.csv', index=False)

# Reading the combined CSV file into a dataframe
df = pd.read_csv('combined_cgiplus.csv')

# Displaying the first few rows of the dataframe
df.head()
```
